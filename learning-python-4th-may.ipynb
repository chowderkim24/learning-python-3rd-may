```python
#identifying outliers
sns.pairplot(df)
#ways of removing outliers 1. scaling 2. normalization  3. dropping outliers from the data 4. standardization
```

```python
N= 10**4
np.random.seed(42)
gaussian_data = np.random.normal(size = N)
skewed_data = np.random.lognormal(size = N)
sns.distplot(gaussian_data)
plt.title("Normally Distributed Data")
```

```python
sns.distplot(skewed_data)
plt.title('lognormal data')
```

```python
#how normalization removes outliers
normalized_data = np.log(skewed_data)
sns.distplot(normalized_data)
plt.title('normalized data')
```

```python
sns.distplot(df['LSTAT'])
plt.title('Original LSTAT')
```

```python
normalized_LSTAT = np.log(df['LSTAT'])
sns.distplot(normalized_LSTAT)
plt.title('normalized LSTAT')
```

```python
#how dropping outliers affects the final outcome: drop every value greater them 30 in LSTAT
df_dropped = df[df['LSTAT'] <= 30]
sns.distplot(df_dropped ['LSTAT'])
plt.title('values greater than 30 dropped')
```

$x_{scaled} = x - x_{min} / x_{max} - x_{min}

```python
zone = df['ZN']
np.min(zone) , np.max(zone)
```

```python
zone = df['ZN']
def min_max_scalar(col):
    scaled_result = (col - min(col))/(max(col) - min(col))
    return scaled_result
scaled_zone = min_max_scalar(zone)
np.min(scaled_zone), np.max(scaled_zone)
```

```python
#categorical data handling
new_df = df.dropna()
new_df.CHAS.unique()
```

```python
#categorical data handling using pandas get dummies function
cat_col = ['CHAS']
encoded_df = pd.get_dummies(data = new_df , columns = cat_col)
encoded_df.head(2)
```

```python
#implementing frequency encoding to handle categorical data
def freq_encoder(df,col):
    fe = df.groupby(col).size()/len(df)
    df[col + "_encoded"] = df[col].map(fe)
    return df
freq_enc_df = freq_encoder(new_df, 'CHAS')
freq_enc_df.head(2)
```

```python
freq_enc_df.CHAS_encoded.unique()
```
